## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

Lane 4: CTR/Engagement Opportunity Scoring.
I frame it primarily as a ranking task because the decision isn't just whether a page is good or bad, it's to provide a ranked list for an editor to look at.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

For the first experiment, features will be calculated from March 1st until March 21st, 2026. The outcome window will be March 22nd until March 31st, 2026.
A page will be marked as future_low_ctr_opportunity when it has sufficient outcome-window impressions and its outcome_window CTR is below the average for pages in a similar position band.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

Precision@50
Of the 50 pages ranked highest for review in held-out data, it measures the proportion that actuall yhave the later low-CTR proxy. I will compare this with the proxy's base rate.

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

In this starter dataframe, one row represents one content page summarized over a trailing 90-day window.

```python
from pathlib import Path
import subprocess
import pandas as pd

REPO_PATH = Path("/content/MLFlyrank")

# Download the repository if this Colab runtime does not have it yet.
if not REPO_PATH.exists():
    subprocess.run(
        [
            "git",
            "clone",
            "--depth",
            "1",
            "https://github.com/esmajasa/MLFlyrank.git",
            str(REPO_PATH),
        ],
        check=True,
    )

DATA_PATH = REPO_PATH / "data/raw/content_refresh_anonymized.csv"

if not DATA_PATH.exists():
    raise FileNotFoundError(f"Dataset not found at: {DATA_PATH}")

df = pd.read_csv(DATA_PATH)

lane_slice = df[
    (df["impressions_90d"] >= 100) &
    (df["avg_position"] > 0)
].copy()

print("Dataset shape:", df.shape)
print("Rows in lane slice:", len(lane_slice))
print("Unique content IDs:", lane_slice["content_id"].nunique())
print("Duplicate content IDs:", lane_slice["content_id"].duplicated().sum())

display(
    lane_slice[
        [
            "content_id",
            "client_id",
            "impressions_90d",
            "ctr",
            "avg_position",
            "position_tier",
            "engagement_rate",
        ]
    ].head()
)
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule could flag every page with CTR below a certain treshold, but CTR depends strongly on average position, impression volume, visibility consistency, and ranking stability, and these signals can interact. A ranking model may combine these signals more flexibly than one threshold.
