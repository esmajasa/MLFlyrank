## 1. My lane (or freestyle) and why

*Name your lane — or say 'freestyle' and describe your own question. One short paragraph: why this one?*

```python
My lane: CTR / Engagement Opportunity Scoring

The goal is to identify visible content pages that may be getting fewer search clicks/weaker engagement than comparable pages.

I chose this lane because it leads to a practical decision: an editor can review a short, prioritized list of pages rather than trying to review the full content inventory. I will begin with CTR as the main signal and use engagement as supporting context. I will compare pages within the same position tier, because ranking position strongly affects CTR.
```

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
Unit of analysis: one row represents one pseudonymized content page in the starter dataset

Decision: Which visible pages should an editor review first for a possible CTR or engagement opportunity?

Output:A ranked review queue. Each selected page will have a score and simple reason codes, such as “high impressions,” “CTR below its position tier average,” or “weak engagement.”

Action: An editor may review the search snippet, title, meta description, intent match, on-page structure, and engagement experience. The editor can then improve the page or choose to monitor it.

Cost of a wrong recommendation: A false positive wastes limited editorial time on a page that does not actually need work. A false negative means a potentially valuable page is not reviewed. Because review time is limited, the queue should prioritize likely opportunities but still require human judgment.

Why data can help: There are many pages and several signals to consider at once, including impressions, CTR, average position, content age, freshness, and engagement. A transparent rule can provide a useful first queue; later, a scoring method may improve that prioritization if it performs better than the simple rule.
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import os
import subprocess

if not os.path.isdir("/content/MLFlyrank"):
    subprocess.run(
        ["git", "clone", "--depth", "1",
         "https://github.com/esmajasa/MLFlyrank.git",
         "/content/MLFlyrank"],
        check=True
    )

os.chdir("/content/MLFlyrank")

import pandas as pd
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Loaded!", df.shape)

visible = df[df["impressions_90d"] >= 100].copy()

print(f"All pages: {len(df):,}")
print(f"Pages with at least 100 impressions: {len(visible):,}")
print(f"Share of pages with at least 100 impressions: {len(visible) / len(df):.1%}")

corr = visible["avg_position"].corr(visible["ctr"])
print(f"Correlation between average position and CTR: {corr:.3f}")

ctr_by_position = (
    visible.groupby("position_tier")["ctr"]
    .agg(["count", "mean"])
    .round(3)
)

print("\nPages and mean CTR by position tier:")
print(ctr_by_position.to_string())
```
